```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.style.use('_mpl-gallery')
sns.set(style='darkgrid')


data=pd.read_csv('job-market(3).csv')

data.head(10)
#Load job market
```

```python
counts = data['Classification'].value_counts()
```

```python
counts
```

```python
plt.figure(figsize=(6,6), dpi=100)

plt.xlabel('Number of Postings')
plt.title('Number of Jobs by Classification')
sns.barplot(y=counts.index, x=counts.values, data=data)
```

```python
salary = data.groupby('LowestSalary')
```

```python
salary.count()
```

```python
salary['HighestSalary'].count()
```

```python
sub_class= salary['SubClassification'].count()
```

```python
#salary ranges plot

plt.figure(figsize=(5,5), dpi=140)
plt.pie(sub_class, labels=sub_class.index, autopct='%1.1f%%');
```
